# Exploratory Data Analysis

## Gender

```python
ax = sns.countplot(x = 'Gender', hue = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', hue = 'Gender', data = sales_gen)
```

**The graph highlights that females not only make up the bulk of buyers but also outspend their male counterparts.**

## Age

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers :
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, hue='Age Group', legend=False)
```

**The graphs demonstrate that the most significant portion of buyers consists of females in the 26-35 age range.**

## State

```python
sales_state = df.groupby (['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders', hue='State', legend=False)
```

```python
sales_state = df.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount', hue='State', legend=False)
```

**The above graphs highlight that the highest order volumes/total sales is dominated by Uttar Pradesh, followed by Maharashtra and then Karnataka  .**

## Marital Status

```python
ax =  sns.countplot(data =df , x = 'Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount' , hue = 'Gender')
```

**From the graphs, it is clear that married women constitute the majority of buyers and have a substantial influence on sales due to their high purchasing power.**

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax =  sns.countplot(data =df , x = 'Occupation', hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount',hue='Occupation')
```

**From the graphs we can see that the buyers predominantly come from the IT, Healthcare, Aviation sectors.**

## Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax =  sns.countplot(data =df , x = 'Product_Category',hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount',hue='Product_Category')
```

**The above graphs shows that the most sold products fall under the Food, Clothing & Apparel and Electronics & Gadgets segments.**

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind = 'bar')
```

# Conclusion:

**Married women aged 26-32 years, from Uttar Pradesh, Maharashtra, and Karnataka, and employed in the IT, Healthcare, and Aviation sectors, are most inclined to purchase products from the Food, Clothing & Apparel, and Electronics & Gadgets categories.**
